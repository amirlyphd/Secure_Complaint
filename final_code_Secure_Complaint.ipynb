```python
# Date: 2025-02-10
# Version: 1.0

"""
Secure Complaint: Advanced  Steganographic Techniques for Secure and Anonymous Audio-Based Complaint Submission

Often hindered by privacy concerns, fear of retaliation, and the complexity of traditional submission
methods the reporting complaint systems in any organization. This paper presents a Secure Audio-Based
 Complaint Submission Framework, designed as the initial phase of any a comprehensive complaint management system.
 The framework ensures the secure mapping of encrypted user details to audio complaints by leveraging advanced steganographic techniques.
 To guarantee confidentiality, user details are first encrypted using AES and further obfuscated through bit reversal and flipping to enhance security.
 The audio complaint undergoes interpolation, doubling the sample count to increase embedding capacity while preserving audio quality.
 The obfuscated data is then embedded into the least significant bits of the interpolated audio samples, ensuring imperceptibility.
 The integrity and quality of the embedded audio are rigorously evaluated using Peak Signal-to-Noise Ratio, Structural Similarity Index Measure, and Bit Error Rate.
 The framework also supports efficient extraction and decryption of user details, with detailed charts provided for transparency.
 Implemented in Python, this system offers a secure, efficient, and anonymous alternative to conventional complaint reporting methods.
 By embedding encrypted user details directly within the complaint audio, the framework ensures strong privacy protection, accurate mapping,
 data integrity, and seamless retrieval, addressing challenges in modern complaint management systems.
"""

# -----------------------------------------------------------------------------
# # Install & import required libraries
# -----------------------------------------------------------------------------
# !pip install pycryptodome
# !pip install scikit-image
# !pip install matplotlib
# !pip install scipy librosa
# !pip install graphviz

import os
import time
import math
import numpy as np
import pandas as pd
import librosa
import librosa.display
import matplotlib.pyplot as plt
from scipy.io import wavfile
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
from skimage.metrics import structural_similarity as ssim
from graphviz import Digraph
from IPython.display import display

# -----------------------------------------------------------------------------
# Configuration and Folder Setup
# -----------------------------------------------------------------------------

COMPLAINT_RECORD_FOLDER = "/content/Complaint_Audio_records"       # Input folder for complaint records (audio)
USER_DETAILS_FOLDER = "/content/user_details"                       # Input folder for user details (text)
OUTPUT_COMPLAINT_RECORD_FOLDER = "/content/Output_Complaint_Audio_records"  # Output folder for stego complaint records
OUTPUT_USER_DETAILS_FOLDER = "/content/Output_user_details"                  # Output folder for extracted user details

# Create output folders if they do not exist
os.makedirs(OUTPUT_COMPLAINT_RECORD_FOLDER, exist_ok=True)
os.makedirs(OUTPUT_USER_DETAILS_FOLDER, exist_ok=True)

# Define the AES encryption key (16 bytes)
ENCRYPTION_KEY = b'Sixteen byte key'

# -----------------------------------------------------------------------------
# Encryption / Decryption Functions for User Details
# -----------------------------------------------------------------------------

def encrypt_user_details(user_details: str, key: bytes) -> bytes:
    """
    Encrypts user details using AES (CBC mode) with the provided key.
    The IV is prepended to the ciphertext.

    Args:
        user_details: The plaintext user details to encrypt.
        key: The encryption key.

    Returns:
        The encrypted user details with the IV prepended.
    """
    cipher = AES.new(key, AES.MODE_CBC)
    encrypted_bytes = cipher.iv + cipher.encrypt(pad(user_details.encode('utf-8'), AES.block_size))
    return encrypted_bytes


def decrypt_user_details(encrypted_bytes: bytes, key: bytes) -> str:
    """
    Decrypts the encrypted user details (with IV prepended) using AES (CBC mode).

    Args:
        encrypted_bytes: The encrypted user details with IV prepended.
        key: The encryption key.

    Returns:
        The decrypted plaintext user details.
    """
    iv = encrypted_bytes[:16]
    cipher = AES.new(key, AES.MODE_CBC, iv)
    decrypted = unpad(cipher.decrypt(encrypted_bytes[16:]), AES.block_size)
    return decrypted.decode('utf-8')

# -----------------------------------------------------------------------------
# Obfuscation / Deobfuscation Functions (for binary conversion)
# -----------------------------------------------------------------------------

def obfuscate_binary(binary_string: str) -> str:
    """
    Obfuscates a binary string by reversing it and flipping each bit.

    Args:
        binary_string: The original binary string.

    Returns:
        The obfuscated binary string.
    """
    return ''.join('1' if bit == '0' else '0' for bit in reversed(binary_string))


def deobfuscate_binary(binary_string: str) -> str:
    """
    Reverses the obfuscation by reversing the binary string and flipping each bit.

    Args:
        binary_string: The obfuscated binary string.

    Returns:
        The original binary string.
    """
    return ''.join('1' if bit == '0' else '0' for bit in reversed(binary_string))

# -----------------------------------------------------------------------------
# Audio Interpolation Function for Complaint Records
# -----------------------------------------------------------------------------

def interpolate_complaint_record(complaint_record: np.ndarray) -> np.ndarray:
    """
    Performs interpolation on the complaint record (audio signal) to double the sample count.
    Original samples are placed at even indices and the average of consecutive samples is
    inserted at odd indices.

    Args:
        complaint_record: The original 1D complaint record (numpy array).

    Returns:
        The interpolated complaint record as a numpy array of type int16.
    """
    norm_record = complaint_record.astype(np.int32) + 32768  # Shift to positive to prevent overflow
    interp_len = norm_record.size * 2 - 1
    interpolated = np.empty(interp_len, dtype=np.int32)

    # Place original samples at even indices
    interpolated[0::2] = norm_record
    # Insert the average of consecutive samples at odd indices
    interpolated[1::2] = (norm_record[:-1] + norm_record[1:]) // 2

    # Clip values to prevent overflow when converting to int16
    interpolated = np.clip(interpolated, 0, 65535)
    return (interpolated - 32768).astype(np.int16)

# -----------------------------------------------------------------------------
# Embedding and Extraction Functions for User Details
# -----------------------------------------------------------------------------

def embed_user_details(complaint_record: np.ndarray, user_details_str: str) -> (np.ndarray, str, float):
    """
    Embeds encrypted user details (provided as a string) into the complaint record.
    The user details are converted to an obfuscated binary string and each bit is embedded into
    the least significant bit (LSB) of the interpolated complaint record samples.

    Args:
        complaint_record: The original complaint record (numpy array).
        user_details_str: The encrypted user details as a string.

    Returns:
        A tuple containing:
          - The stego complaint record (numpy array).
          - The binary string of the embedded user details.
          - The time taken for embedding.
    """
    start_time = time.time()
    # Convert each character to its 8-bit binary representation
    binary_str = ''.join(format(ord(c), '08b') for c in user_details_str)
    # Obfuscate the binary string
    user_details_bits = obfuscate_binary(binary_str)

    # Interpolate the complaint record
    interpolated_record = interpolate_complaint_record(complaint_record)

    # Check if there are enough samples for embedding
    if len(user_details_bits) > len(interpolated_record):
        raise ValueError("Not enough complaint record samples for embedding the user details.")

    # Embed each bit into the LSB of the interpolated complaint record samples
    for i, bit in enumerate(user_details_bits):
        interpolated_record[i] = (interpolated_record[i] & ~1) | int(bit)

    end_time = time.time()
    return interpolated_record, user_details_bits, end_time - start_time


def extract_user_details(stego_record: np.ndarray, details_length: int) -> (str, str, float):
    """
    Extracts embedded user details from the stego complaint record by reading the LSBs.
    The number of characters in the original encrypted user details is provided by details_length.

    Args:
        stego_record: The complaint record containing the embedded user details (numpy array).
        details_length: The number of characters in the encrypted user details.

    Returns:
        A tuple containing:
          - The extracted encrypted user details as a string.
          - The binary string extracted from the LSBs.
          - The time taken for extraction.
    """
    start_time = time.time()
    total_bits = details_length * 8
    extracted_bits_array = stego_record[:total_bits] & 1
    extracted_bits = ''.join(extracted_bits_array.astype(str))

    # Reverse obfuscation to get the original binary string
    recovered_bits = deobfuscate_binary(extracted_bits)
    # Convert every 8 bits into a character
    extracted_chars = ''.join(
        chr(int(recovered_bits[i:i+8], 2)) for i in range(0, len(recovered_bits), 8)
    )

    end_time = time.time()
    return extracted_chars, extracted_bits, end_time - start_time

# -----------------------------------------------------------------------------
# Quality Metrics Functions
# -----------------------------------------------------------------------------

def calculate_metrics(original_record: np.ndarray, stego_record: np.ndarray) -> (float, float):
    """
    Calculates quality metrics (PSNR and SSIM) between the original and stego complaint records.
    For SSIM, the 1D signals are converted to 2D arrays by tiling them into 3 rows.

    Args:
        original_record: The original complaint record.
        stego_record: The stego complaint record.

    Returns:
        A tuple containing:
          - The PSNR value.
          - The SSIM value.
    """
    # Ensure both records are of equal length
    min_length = min(len(original_record), len(stego_record))
    original_record = original_record[:min_length]
    stego_record = stego_record[:min_length]

    mse = np.mean((original_record - stego_record) ** 2)
    psnr_value = 20 * np.log10(np.max(np.abs(original_record)) / np.sqrt(mse)) if mse > 0 else float('inf')

    # Convert to 2D arrays (3 rows) for SSIM calculation
    original_2d = np.tile(original_record.astype(np.float32), (3, 1))
    stego_2d = np.tile(stego_record.astype(np.float32), (3, 1))

    data_range = original_2d.max() - original_2d.min()
    if data_range == 0:
        data_range = 1e-10

    ssim_value = ssim(original_2d, stego_2d, data_range=data_range, win_size=3)

    return psnr_value, ssim_value

# -----------------------------------------------------------------------------
# Plotting Function for Complaint Record Comparison
# -----------------------------------------------------------------------------

def plot_complaint_record_comparison(original_path: str, stego_path: str):
    """
    Plots the waveforms and spectrograms of the original and stego complaint records.

    Args:
        original_path: Path to the original complaint record file.
        stego_path: Path to the stego complaint record file.
    """
    try:
        sr_orig, record_orig = wavfile.read(original_path)
        sr_stego, record_stego = wavfile.read(stego_path)
    except Exception as e:
        print(f"Error reading complaint record files: {e}")
        return

    plt.figure(figsize=(12, 8))
    plt.subplot(2, 1, 1)
    plt.title("Original Complaint Record Waveform")
    plt.plot(record_orig)

    plt.subplot(2, 1, 2)
    plt.title("Stego Complaint Record Waveform")
    plt.plot(record_stego)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(12, 8))
    plt.subplot(2, 1, 1)
    plt.title("Original Complaint Record Spectrogram")
    D_orig = librosa.amplitude_to_db(np.abs(librosa.stft(record_orig.astype(float))), ref=np.max)
    librosa.display.specshow(D_orig, sr=sr_orig, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')

    plt.subplot(2, 1, 2)
    plt.title("Stego Complaint Record Spectrogram")
    D_stego = librosa.amplitude_to_db(np.abs(librosa.stft(record_stego.astype(float))), ref=np.max)
    librosa.display.specshow(D_stego, sr=sr_stego, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')

    plt.tight_layout()
    plt.show()

# -----------------------------------------------------------------------------
# Processing Function for Complaint Records and User Details
# -----------------------------------------------------------------------------

def process_complaint_records():
    """
    Processes complaint records (audio files) and user details (text files) by pairing one set of user details
    per complaint record. For each pair, the following steps are executed:
      1. Read the complaint record and corresponding user details.
      2. Encrypt the user details.
      3. Embed the encrypted user details into the complaint record.
      4. Save the stego complaint record.
      5. Extract the user details from the stego complaint record and decrypt them.
      6. Compute quality metrics (PSNR, SSIM, and BER).

    Returns:
        A list of metrics for each processed complaint record and user details pair.
    """
    metrics_data = []

    # Get sorted lists of complaint record files and user details text files
    complaint_files = sorted([f for f in os.listdir(COMPLAINT_RECORD_FOLDER) if f.endswith(".wav")])
    details_files = sorted([f for f in os.listdir(USER_DETAILS_FOLDER) if f.endswith(".txt")])

    if not complaint_files:
        print("No complaint record files found in the specified folder.")
        return metrics_data
    if not details_files:
        print("No user details files found in the specified folder.")
        return metrics_data

    # Warn if there are more complaint records than user details files
    if len(complaint_files) > len(details_files):
        print("Warning: Fewer user details files than complaint records. The first user details file will be used for remaining complaint records.")

    # Process each complaint record, pairing it with a user details file
    for i, complaint_file in enumerate(complaint_files):
        complaint_path = os.path.join(COMPLAINT_RECORD_FOLDER, complaint_file)
        try:
            sample_rate, complaint_record = wavfile.read(complaint_path)
        except Exception as e:
            print(f"Error reading {complaint_file}: {e}")
            continue
        complaint_record = complaint_record.astype(np.int16)

        # Select user details: if there are fewer details files, reuse the first one
        details_file = details_files[i] if i < len(details_files) else details_files[0]
        details_path = os.path.join(USER_DETAILS_FOLDER, details_file)
        try:
            with open(details_path, "r") as f:
                user_details = f.read().strip()
        except Exception as e:
            print(f"Error reading {details_file}: {e}")
            continue

        # Encrypt the user details
        try:
            encrypted_details_bytes = encrypt_user_details(user_details, ENCRYPTION_KEY)
        except Exception as e:
            print(f"Encryption error for {details_file}: {e}")
            continue

        # Convert encrypted details to a string using Latin-1 encoding
        embedded_details_str = encrypted_details_bytes.decode('latin1')

        # Embed the encrypted user details into the complaint record
        try:
            stego_record, details_bits, embed_time = embed_user_details(complaint_record, embedded_details_str)
        except ValueError as ve:
            print(f"Embedding error in {complaint_file}: {ve}")
            continue

        stego_record_filename = f"stego_{complaint_file}"
        stego_record_path = os.path.join(OUTPUT_COMPLAINT_RECORD_FOLDER, stego_record_filename)
        try:
            wavfile.write(stego_record_path, sample_rate * 2, stego_record)
        except Exception as e:
            print(f"Error writing stego complaint record {stego_record_path}: {e}")
            continue

        # Extract the user details from the stego complaint record
        extracted_details_str, extracted_bits, extract_time = extract_user_details(stego_record, len(embedded_details_str))
        try:
            decrypted_details = decrypt_user_details(extracted_details_str.encode('latin1'), ENCRYPTION_KEY)
        except Exception as e:
            decrypted_details = f"Decryption failed: {e}"

        output_details_path = os.path.join(OUTPUT_USER_DETAILS_FOLDER, f"extracted_{details_file}")
        try:
            with open(output_details_path, "w") as f:
                f.write(decrypted_details)
        except Exception as e:
            print(f"Error writing extracted user details to {output_details_path}: {e}")

        # Calculate quality metrics (PSNR, SSIM)
        psnr_value, ssim_value = calculate_metrics(complaint_record, stego_record)
        # Calculate Bit Error Rate (BER)
        errors = sum(o != e for o, e in zip(details_bits, extracted_bits))
        ber_value = errors / len(details_bits) if details_bits else None

        metrics_data.append([
            complaint_file, details_file, psnr_value, ber_value, ssim_value, embed_time, extract_time
        ])

    return metrics_data

# -----------------------------------------------------------------------------
# Flowchart Generation Functions
# -----------------------------------------------------------------------------

def generate_flowcharts():
    """
    Generates flowcharts for the embedding and extraction processes.
    The flowcharts are saved as PNG images.
    """
    # Embedding Process Flowchart
    embedding_flow = Digraph("Embedding_Process", format="png")
    embedding_flow.attr(size="12,8", fontsize="12")

    embedding_flow.node("Start", "Start", shape="ellipse", style="filled", fillcolor="lightblue")
    embedding_flow.node("ReadRecord", "Read Complaint Record", shape="parallelogram", style="filled", fillcolor="lightyellow")
    embedding_flow.node("ReadDetails", "Read User Details", shape="parallelogram", style="filled", fillcolor="lightyellow")
    embedding_flow.node("EncryptDetails", "Encrypt User Details (AES)", shape="rectangle", style="filled", fillcolor="lightgreen")
    embedding_flow.node("ConvertBinary", "Convert User Details to Binary & Obfuscate", shape="rectangle", style="filled", fillcolor="lightgreen")
    embedding_flow.node("Interpolate", "Interpolate Complaint Record Samples", shape="rectangle", style="filled", fillcolor="lightgreen")
    embedding_flow.node("EmbedBits", "Embed User Details Bits into Complaint Record (LSB)", shape="rectangle", style="filled", fillcolor="lightgreen")
    embedding_flow.node("SaveStego", "Save Stego Complaint Record", shape="parallelogram", style="filled", fillcolor="lightyellow")
    embedding_flow.node("PlotWaveforms", "Plot Complaint Record Waveforms/Spectrograms", shape="rectangle", style="filled", fillcolor="lightpink")
    embedding_flow.node("CalculateMetrics", "Calculate Quality Metrics (PSNR, BER, SSIM)", shape="rectangle", style="filled", fillcolor="lightpink")
    embedding_flow.node("End", "End", shape="ellipse", style="filled", fillcolor="lightblue")

    embedding_flow.edge("Start", "ReadRecord", label="Input")
    embedding_flow.edge("Start", "ReadDetails", label="Input")
    embedding_flow.edge("ReadDetails", "EncryptDetails", label="Encrypt Details")
    embedding_flow.edge("EncryptDetails", "ConvertBinary", label="Convert to Binary")
    embedding_flow.edge("ReadRecord", "Interpolate", label="Prepare Record")
    embedding_flow.edge("Interpolate", "EmbedBits", label="Modify Samples")
    embedding_flow.edge("ConvertBinary", "EmbedBits", label="Embed Bits")
    embedding_flow.edge("EmbedBits", "SaveStego", label="Save Output")
    embedding_flow.edge("SaveStego", "PlotWaveforms", label="Visualize")
    embedding_flow.edge("PlotWaveforms", "CalculateMetrics", label="Evaluate Quality")
    embedding_flow.edge("CalculateMetrics", "End", label="Complete")

    embedding_flow.render("/content/charts/embedding_flowchart")
    print("Embedding process flowchart saved at /content/charts/embedding_flowchart.png")

    # Extraction Process Flowchart
    extraction_flow = Digraph("Extraction_Process", format="png")
    extraction_flow.attr(size="12,8", fontsize="12")

    extraction_flow.node("Start", "Start", shape="ellipse", style="filled", fillcolor="lightblue")
    extraction_flow.node("ReadStego", "Read Stego Complaint Record", shape="parallelogram", style="filled", fillcolor="lightyellow")
    extraction_flow.node("ExtractBits", "Extract LSBs", shape="rectangle", style="filled", fillcolor="lightgreen")
    extraction_flow.node("Deobfuscate", "Deobfuscate Binary", shape="rectangle", style="filled", fillcolor="lightgreen")
    extraction_flow.node("ConvertText", "Convert Binary to User Details Text", shape="rectangle", style="filled", fillcolor="lightgreen")
    extraction_flow.node("DecryptDetails", "Decrypt User Details (AES)", shape="rectangle", style="filled", fillcolor="lightgreen")
    extraction_flow.node("SaveDetails", "Save Retrieved User Details", shape="parallelogram", style="filled", fillcolor="lightyellow")
    extraction_flow.node("AnalyzeBER", "Analyze Bit Error Rate (BER)", shape="rectangle", style="filled", fillcolor="lightpink")
    extraction_flow.node("PlotComparison", "Plot Complaint Record Waveforms/Spectrograms", shape="rectangle", style="filled", fillcolor="lightpink")
    extraction_flow.node("End", "End", shape="ellipse", style="filled", fillcolor="lightblue")

    extraction_flow.edge("Start", "ReadStego", label="Input")
    extraction_flow.edge("ReadStego", "ExtractBits", label="Extract Bits")
    extraction_flow.edge("ExtractBits", "Deobfuscate", label="Deobfuscate")
    extraction_flow.edge("Deobfuscate", "ConvertText", label="Binary to Text")
    extraction_flow.edge("ConvertText", "DecryptDetails", label="Decrypt Details")
    extraction_flow.edge("DecryptDetails", "SaveDetails", label="Save Output")
    extraction_flow.edge("SaveDetails", "AnalyzeBER", label="Error Analysis")
    extraction_flow.edge("AnalyzeBER", "PlotComparison", label="Visualize")
    extraction_flow.edge("PlotComparison", "End", label="Complete")

    extraction_flow.render("/content/charts/extraction_flowchart")
    print("Extraction process flowchart saved at /content/charts/extraction_flowchart.png")

# -----------------------------------------------------------------------------
# Main Execution
# -----------------------------------------------------------------------------

def main():
    """
    Main execution function.
    Processes the complaint records and user details, displays the computed metrics,
    generates flowcharts, and plots an example complaint record comparison.
    """
    results = process_complaint_records()
    if results:
        df = pd.DataFrame(
            results,
            columns=["Complaint Record File", "User Details File", "PSNR", "BER", "SSIM", "Embed Time", "Extract Time"]
        )

        # Extract the numeric part from the complaint record file names for sorting
        df['Complaint Number'] = df['Complaint Record File'].str.extract(r'complaint_audio_(\d+)').astype(int)

        # Sort by the extracted numeric part
        df_sorted = df.sort_values(by='Complaint Number')

        # Drop the helper column
        df_sorted = df_sorted.drop(columns=['Complaint Number'])

        # Display the sorted DataFrame
        display(df_sorted)

        # Save the sorted DataFrame to an Excel file
        df_sorted.to_excel("complaint_metrics.xlsx", index=False)
        print("Data saved to 'complaint_metrics.xlsx'")
    else:
        print("No files were processed.")

    # Example: Plot comparison for a selected complaint record pair (if available)
    original_record_file = os.path.join(COMPLAINT_RECORD_FOLDER, "/content/Complaint_Audio_records/complaint_audio_16.wav")
    stego_record_file = os.path.join(OUTPUT_COMPLAINT_RECORD_FOLDER, "/content/Output_Complaint_Audio_records/stego_complaint_audio_16.wav")
    if os.path.exists(original_record_file) and os.path.exists(stego_record_file):
        plot_complaint_record_comparison(original_record_file, stego_record_file)
    else:
        print("Original complaint record or stego complaint record does not exist for comparison.")

    # Generate flowcharts for embedding and extraction processes
    generate_flowcharts()


if __name__ == "__main__":
    main()
```
